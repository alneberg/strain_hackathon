```python
def conditional_posterior(c, g, L, eta, pi_s_g, n, a, scale):
    """ 
    c is the contig currently considered unkown
    
    g is the genome currently considered
    
    L is a vector containing all contig lengths
    L[c] is the length of contig c
    
    eta is the copy number for each contig in the current genome
    eta[g, c] is the copy number of contig c in genome g
    
    pi_s_g is the relative abundance for each sample and genome
    pi_s_g[s,g] is the relative abundance of genome g in sample s
    
    n is the number of reads mapping to each contig at each site
    n[c,s] is the number of reads mapping to contig c at site s
    """
    inner_sum = L*np.dot(pi_s_g,eta)
    lognuN = np.log(inner_sum[:,c]) - \
                np.log(np.sum(inner_sum))
    genome_lengths = np.sum(eta*L, axis=1)
    return np.sum(lognuN*n[c,:] - np.exp(lognuN) - scipy.special.gammaln(n[c,:]+1)) + np.sum(gamma.logpdf(genome_lengths, a, scale=scale))
```

```python
# After an original at https://gist.github.com/harrymvr/3991982
import numpy as np

def log_add(x, y):
    maximum = max(x,y)
    minimum = min(x,y)
    if(abs(maximum - minimum) > 30):
        # the difference is too small, return the just the maximum
        return maximum
    return maximum + np.log(1 + np.exp(minimum - maximum))
 
# Takes as input a log probability vector. The base of the logarithm is 2.
# Returns a sampled position according to the corresponding log probabilities
# Uses the formula from
# http://blog.smola.org/post/987977550/log-probabilities-semirings-and-floating-point-numbers
def sample_from_log_prob(A):
    print A
    C = [A[0]]
    for a in A[1:]:
        C.append(log_add(C[-1], a))
    C_pos = [ -c for c in reversed(C)]
    r = np.log(np.random.uniform())
    print r
    pos = np.searchsorted(C_pos,-r)
    print len(C) - pos
    return len(C) - pos

def sample_from_non_norm_log_prob(A):
    # If the sum dominated by one value, choose that one
    B = np.sort(A)
    maximum = B[-1]
    second_largest = B[-2]
    if abs(maximum - second_largest) > 100:
        return A.argmax()
    # Create a normalized probability vector
    C = [A[0]]
    for a in A[1:]:
        C.append(log_add(C[-1], a))
    r = np.log(np.random.uniform())
    D = C - C[-1]
    pos = np.searchsorted(D, r)
    return pos
```

```python
# Trying out a gibbs sampler for eta.
# Pi is fixed at a correct level.
N = 10 # Number of iterations in Gibbs sampling
G = 2
C = 2
eta_prob = np.ones((G,C,10))
L = np.array([1000, 2000])
eta = np.array([[1,2], [2,0]])
eta_saved = eta.copy()
pi_s_g = np.array([[0.65, 0.35], [0.70, 0.30]])
n = np.array([[200, 50], [100, 80]])
a = 1000. # Gives a nice shape of the density
mean_length = 2.5e3
scale = mean_length/a
old_prob = -np.inf
prec = 1e-4
for i in xrange(N):
    for g in xrange(G):
        for c in xrange(C):
            old_val = eta[g,c]
            for eta_val in range(10):
                eta[g,c] = eta_val
                eta_prob[g, c, eta_val] = conditional_posterior(c, g, L, eta, pi_s_g, n, a, scale)
            eta[g,c] = sample_from_non_norm_log_prob(eta_prob[g,c,:])
            if  old_val != eta[g,c]:
                print i, "Changing eta {0} {1} from {2} to {3}".format(g, c, old_val, eta[g,c])
            

    new_prob = conditional_posterior(c, g, L, eta, pi_s_g, n, a, scale)
    if np.abs(new_prob - old_prob) < prec:
        break
    else:
        old_prob = new_prob
print eta
```
