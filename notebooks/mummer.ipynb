# Try to mimic Chris procedure

```python
MAX_OVERLAP = 0.4

def _large_overlap(mapped_start, mapped_end, start_pos, end_pos):
    length = float(end_pos - start_pos + 1)
    #Check whether any overlap at all:
    if mapped_start < start_pos and mapped_end > start_pos:
        # Check if overlap is more than MAX_OVERLAP
        overlap = mapped_end - start_pos + 1
        if overlap / length > MAX_OVERLAP:
            return True
    return False

def too_large_overlap(mapped_regions, start_pos, end_pos):
    """The mapped_regions is a list of tuples with start positions and end positions."""
    # Make sure the direction of the contig is ok
    if end_pos < start_pos:
        start_pos, end_pos = end_pos, start_pos
    for mapped_start, mapped_end in mapped_regions:
        if _large_overlap(mapped_start, mapped_end, start_pos, end_pos):
            return True
    return False

def return_overlap(DF, idr, q_s1, q_e1):
    overlaps = set()
    df_subset = DF[DF.IDR == idr]
    for i, row in df_subset.iterrows():
            
        if row.S1 > row.E1:
            s1, e1 = row.E1, row.S1
        else:
            s1, e1 = row.S1, row.E1
        if _large_overlap(s1, e1, q_s1, q_e1):
            overlaps.add(row.IDQ)
    return overlaps
```

```python
ALGN_MIN_LEN = 90
AVG_PRC_ID = 0.99
contigs_without_long_algn = set()
# Mapped regions is to keep track of contigs overlapping in the reference
mapped_regions = defaultdict(list)
DF = nucmer_result
pre_keepers = defaultdict(list)
keepers = set()
ambigous = set()
overlaps = []
skipped_due_to_overlap = set()
# For each mapping only consider the ones of proper length and 
# percentage identity. 
# Among these, only add alignments if overlapping is short enough.
# Last of all check if contig has more than one reference alignment, remove those.
# Use the ones that are still left.
#
# Mind the overlaps and duplications.
for i, row in DF.iterrows():
    # Check whether this reference is among the ones
    # with a good alignment length
    cov_q = row.COV_Q
    if cov_q < ALGN_MIN_LEN:
        continue

    # Filter on the average prc identity:
    prc_id = row.PRC_IDY / 100.0
    if prc_id < AVG_PRC_ID:
        continue
    
    if row.S1 > row.E1:
        s1, e1 = row.E1, row.S1
    else:
        s1, e1 = row.S1, row.E1
    
    if too_large_overlap(mapped_regions[row.IDR], s1, e1):
        skipped_due_to_overlap.add(row.IDQ)
        overlaps.append((row.IDQ, return_overlap(DF, row.IDR, s1, e1)))
        continue

    mapped_regions[row.IDR].append((s1, e1))
    # This alignment is not overlapping and 
    pre_keepers[row.IDQ].append(row.IDR)


# Filter on multiple reference matches:
for idq, references in pre_keepers.iteritems():
    number_refs = len(references)
    if number_refs > 1:
        ambigous.add(idq)
        continue
    keepers.add(idq)
```
